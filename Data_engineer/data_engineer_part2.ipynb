#### Prescribers - For Practice Level Prescribing Users (PLP)
#### Observe que a classe Prescribers está ligado a prática profissional

```python
df.columns
```

```python
df_prescribers = pd.DataFrame(df, columns=['REGIONAL_OFFICE_NAME', 'REGIONAL_OFFICE_CODE',
                                           'ICB_NAME', 'ICB_CODE', 'PCO_NAME', 'PCO_CODE',
                                           'PRACTICE_NAME','PRACTICE_CODE','BNF_CODE',
                                           'BNF_DESCRIPTION', 'CHEMICAL_SUBSTANCE_BNF_DESCR',
                                           'QUANTITY','ITEMS','ADQUSAGE','NIC','ACTUAL_COST'])
```

```python
df_prescribers
```

```python
# Converte o Arquivo para formato parquet totalizado dos meses maio, junho e julho do EPD.
df_prescribers.to_parquet('/Jupyter/2RP/prescribers_epd_202205_06_07.parquet')
```
